```python
def create_dic(df):
    # Define the columns (adjust as needed)
    lemma_cols = [
            'gorman_lemma_1',
    'gorman_lemma_2',
    'gorman_lemma_3',
    'gorman_lemma_4',
    'gorman_lemma_5',
    'gorman_lemma_6',
    'gorman_lemma_7',
    'gorman_lemma_8',
    'gorman_lemma_9',
    'gorman_lemma_10',
    'gorman_lemma_11',
    'gorman_lemma_12',
    'gorman_lemma_13',
    'gorman_lemma_14',
    'gpt_lemma_1',
    'gpt_lemma_2',
    'gpt_lemma_3',
    'gpt_lemma_4',
    'gpt_lemma_5',
    'gpt_lemma_6',
    'gpt_lemma_7',
    'gpt_lemma_8',
    'gpt_lemma_9',
    'gpt_lemma_10',
    'gpt_lemma_11',
    'gpt_lemma_12',
    'gpt_lemma_13',
    'gpt_lemma_14',
            'lemma_pedalion_1', 'lemma_pedalion_2', 'lemma_pedalion_3',
            'lemma_pedalion_4', 'lemma_pedalion_5', 'lemma_pedalion_6',
            'lemma_pedalion_7', 'lemma_pedalion_8', 'lemma_pedalion_9',
            'lemma_pedalion_10', 'lemma_pedalion_11', 'lemma_pedalion_12',
            'lemma_pedalion_13', 'lemma_pedalion_14', 'lemma_pedalion_15',
            'lemma_pedalion_16', 'lemma_pedalion_17'
        ]
    postag_cols = ['postag_1',
    'postag_2',
    'postag_3',
    'postag_4',
    'postag_5',
    'postag_6',
    'postag_7',
    'postag_8',
    'postag_9',
    'postag_10',
    'postag_11',
    'postag_12',
    'postag_13',
    'postag_14',
    'postag_15',
    'postag_16',
    'postag_17',
    'postag_18',
    'postag_19']

    # Dictionary to store consolidated data per @form
    form_dict = {}
    for _, row in df.iterrows():
        form = row['@form']
        if form not in form_dict:
            form_dict[form] = {"lemmas": set(), "postags": set()}
        # Gather lemma values (avoiding missing values)
        for col in lemma_cols:
            if col in df.columns:
                value = row[col]
                if pd.notnull(value):
                    form_dict[form]["lemmas"].add(value)
        # Gather postag values similarly
        for col in postag_cols:
            if col in df.columns:
                value = row[col]
                if pd.notnull(value):
                    form_dict[form]["postags"].add(value)
    return form_dict

def merge_list_of_dicts(dict_list):
    merged = {}
    for d in dict_list:
        for form, values in d.items():
            if form in merged:
                merged[form]['lemmas'].update(values['lemmas'])
                merged[form]['postags'].update(values['postags'])
            else:
                merged[form] = {
                    'lemmas': set(values['lemmas']),
                    'postags': set(values['postags'])
                }
    return merged
```

```python
list_dir = os.listdir('correzione lemmi')
lis = [pd.read_csv('correzione lemmi/'+i) for i in list_dir]
dic_lis = []
for df in lis:
    dic = create_dic(df)
    dic_lis.append(dic)
merged_dic = merge_list_of_dicts(dic_lis)  
```

```python
def change_lemma(df,dic):
    count = 0
    count_pt = 0
    for idx, row in df.iterrows():
        form = row['@form']
        # Check if form exists in the dictionary
        if form in dic:
            valid_lemmas = dic[form]['lemmas']
            # Check if the current lemma is not in the valid lemmas
            if row['@lemma'] not in valid_lemmas:
                # Pick the first lemma (convert to list for ordering, or use next(iter(valid_lemmas)) if order doesn't matter)
                if len(list(valid_lemmas))>0:
                    new_lemma = list(valid_lemmas)[0]
                    df.at[idx, '@lemma'] = new_lemma
                    count += 1
            else:
                if len(list(valid_lemmas))>0:
                    new_lemma = row['@lemma']
            valid_postags = dic[form]['postags']
            # Check if the current lemma is not in the valid lemmas
            if row['@postag'] not in valid_postags:
                # Pick the first lemma (convert to list for ordering, or use next(iter(valid_lemmas)) if order doesn't matter)
                if len(list(valid_postags))>0:
                    '''for pv in valid_postags:
                        differenze = [i for i,t in enumerate(row['@postag']) if t != pv[i]]
                        if differenze:
                            dif_list_postag.append((row['@form'], row['@postag'], pv))'''
                    new_postag = row['@postag']
                    df.at[idx, '@postag'] = new_postag
                    count_pt += 1        
            

    return count,count_pt,df
```

```python
def df_to_dict(df):
    # Identify the columns for lemmas and postags
    lemma_columns = [col for col in df.columns if col.startswith('lemma_')]
    postag_columns = [col for col in df.columns if col.startswith('postag_')]
    
    result = {}
    
    # Iterate over each row in the DataFrame
    for _, row in df.iterrows():
        form = row['@form']
        
        # Initialize the dictionary entry if the form is encountered for the first time
        if form not in result:
            result[form] = {'lemmas': set(), 'postags': set()}
        
        # Add non-null and non-empty lemma values from the corresponding columns
        for col in lemma_columns:
            lemma = row[col]
            if pd.notnull(lemma) and lemma != "":
                result[form]['lemmas'].add(lemma)
        
        # Add non-null and non-empty postag values from the corresponding columns
        for col in postag_columns:
            postag = row[col]
            if pd.notnull(postag) and postag != "":
                result[form]['postags'].add(postag)
                
    return result
```

```python
import os
import pandas as pd
from collections import defaultdict

# Directory contenente i dati
base_dir = 'output_gorman'

# Dizionario dei lemmi puliti (già definito)
# dizionario_pulito = ...

# Statistiche da raccogliere
lemma_counts = defaultdict(lambda: defaultdict(int))  # {form: {lemma: count}}
total_counts = defaultdict(int)  # {form: total_occurrences}
not_found_forms = set()

# Scansiona tutti i CSV
for root, dirs, files in os.walk(base_dir):
    for file in files:
        if file.endswith('.csv'):
            filepath = os.path.join(root, file)
            try:
                df = pd.read_csv(filepath)
                if '@form' in df.columns and '@lemma' in df.columns:
                    for _, row in df.iterrows():
                        form = row['@form']
                        lemma = row['@lemma']
                        if pd.isna(form) or pd.isna(lemma):
                            continue
                        total_counts[form] += 1
                        lemma_counts[form][lemma] += 1
                else:
                    print(f"Skipping {filepath}, missing expected columns.")
            except Exception as e:
                print(f"Errore leggendo {filepath}: {e}")

# Preparazione della tabella
records = []
for form in total_counts:
    if form in dizionario_pulito:
        lemmi_diz = set(dizionario_pulito[form])
        tot = total_counts[form]
        count_match = sum(count for lemma, count in lemma_counts[form].items() if lemma in lemmi_diz)
        records.append({
            'form': form,
            'lemmi_originari': list(lemma_counts[form].keys()),
            'lemmi_dizionario': list(lemmi_diz),
            'match_ratio': f"{count_match}/{tot}"
        })
    else:
        not_found_forms.add(form)

# Crea la tabella finale
df_final = pd.DataFrame(records)

# Mostra un'anteprima
print(df_final.head())

# Salva se vuoi
# df_final.to_csv('tabella_confronto_lemmi.csv', index=False)

# Forme non trovate nel dizionario
print("\nNumero di forme non trovate nel dizionario:", len(not_found_forms))
# Salva se vuoi
# with open("forme_non_trovate.txt", "w", encoding="utf-8") as f:
#     for form in sorted(not_found_forms):
#         f.write(form + "\n")
```

### Tabella count

```python
import os
import pandas as pd
from collections import defaultdict

# Directory principale
base_dir = 'output_gorman'

# Inizializzazione dei contenitori
form_counts = defaultdict(int)
form_lemmas = defaultdict(set)
form_postags = defaultdict(set)

# Scansione ricorsiva dei CSV
for root, dirs, files in os.walk(base_dir):
    for file in files:
        if file.endswith('.csv'):
            filepath = os.path.join(root, file)
            try:
                df = pd.read_csv(filepath)
                if '@form' in df.columns and '@lemma' in df.columns and '@postag' in df.columns:
                    for _, row in df.iterrows():
                        form = row['@form']
                        lemma = row['@lemma']
                        postag = row['@postag']
                        if pd.notna(form):
                            form_counts[form] += 1
                            if pd.notna(lemma):
                                form_lemmas[form].add(lemma)
                            if pd.notna(postag):
                                form_postags[form].add(postag)
                else:
                    print(f"Colonne mancanti in {filepath}")
            except Exception as e:
                print(f"Errore leggendo {filepath}: {e}")

# Costruzione della tabella finale
records = []
for form in form_counts:
    records.append({
        '@form': form,
        'frequenza': form_counts[form],
        'lemmi': sorted(form_lemmas[form]),
        'n_lemmi': len(form_lemmas[form]),
        'postag': sorted(form_postags[form]),
        'n_postag': len(form_postags[form])
    })

df_summary = pd.DataFrame(records)

# Mostra un'anteprima
print(df_summary.head())
```
