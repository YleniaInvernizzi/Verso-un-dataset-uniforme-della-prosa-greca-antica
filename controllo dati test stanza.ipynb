# funzione per controllare che i dati di train non siano nel test

```python
def check_collu(file_path):
    sent_ids = set()
    sent_texts = set()
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            if line.startswith('# sent_id ='):
                sent_id = line.strip().split('=', 1)[1].strip()
                sent_ids.add(sent_id)
            elif line.startswith('# text ='):
                sent_text = line.strip().split('=', 1)[1].strip()
                sent_texts.add(sent_text)
    return sent_ids, sent_texts

dev_file = 'C:/Users/lcolo/Desktop/TESI/CODICI DEFINITIVI DIZ UNICO-2/GORMAN/train_dev_test/grc_gorman-ud-dev.conllu' #metti il tuo path intero
train_file = 'C:/Users/lcolo/Desktop/TESI/CODICI DEFINITIVI DIZ UNICO-2/GORMAN/train_dev_test/grc_gorman-ud-train.conllu'
test_file = 'C:/Users/lcolo/Desktop/TESI/CODICI DEFINITIVI DIZ UNICO-2/GORMAN/train_dev_test/grc_gorman-ud-test.conllu'

dev_ids, dev_texts = check_collu(dev_file)
train_ids, train_texts = check_collu(train_file)
test_ids, test_texts = check_collu(test_file)
```

```python
all_ids = [("dev", dev_ids), ("train", train_ids), ("test", test_ids)]
all_texts = [("dev", dev_texts), ("train", train_texts), ("test", test_texts)]
```

```python
def find_overlap(name_set_pairs):
    seen = {}
    overlaps = set()
    for name, items in name_set_pairs:
        for item in items:
            if item in seen:
                overlaps.add((item, seen[item], name))
            else:
                seen[item] = name
    return overlaps
```

```python
id_overlaps = find_overlap(all_ids)
print(id_overlaps)
text_overlaps = find_overlap(all_texts)
print(text_overlaps)
```
