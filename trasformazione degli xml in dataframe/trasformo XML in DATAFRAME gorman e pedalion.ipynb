```python
xml_table_final_listrelation = {}
xml_list = os.listdir('C:/Users/lcolo/Desktop/TESI/CODICI DEFINITIVI DIZ UNICO-2/trasformazione degli xml in dataframe/xml_pedalion') #così definisco xml_list come la cartella con tutti gli xml
#questo mi serve per creare un dizionario, nel dizionario voglio mettere tutte le tabelle.
#quali sono le chiavi? Le chiavi stanno nelle quadre. In questo caso le mie chiavi sono tutti gli xml. i valori corrispondenti a queste chiavi sono
#tutte le frasi degli xml. Nella lista vuota mi aspetto di avere le frasi di tutti gli xml. 
for xml_file in xml_list: #xml_file è come i, è ciascun file generico xml, è la mia variabile provvisoria, lo chiamo xml_file per avere un codice parlante.
    xml_table_final_listrelation[xml_file] = [] #ho creato un dizionario che ha per chiave xml_file, cioè i, cioè ogni singolo file xml. ad ogni chiave deve corrispondere 
    #un valore. il mio valore è la lista delle tabelle, cioè tutte le frasi messe in modalità tabellare di ogni singolo xml. 
    #tra quadre che gli dico che quella è la chiave del dizionario se voglio ottenere le tabelle delle frasi di quell'xml. 
    with open('C:/Users/lcolo/Desktop/TESI/CODICI DEFINITIVI DIZ UNICO-2/trasformazione degli xml in dataframe/xml_pedalion/'+xml_file,'r',encoding='utf-8') as f: #+ sto aprendo la cartella madre xmlpedalion e dentro prendo l'xml corrente che cambia ad ogni iterazione del ciclo.
        print(xml_file)
        data = f.read()
    xml_dict = xmltodict.parse(data) #ho trasformato data in dizionario.
 
    for i in range(len(xml_dict['treebank']['sentence'])): 
        #range è un metodo di Python che mi permette di trasformare un numero in una lista di numeri che va da 0 a quel numero -1.
        #se scrivo range(5)
        # range di 5 va da 0 a 4.
        #xml_dict è in quel momento uno specifico xml. 
        #in ogni xml ci sono delle frasi, quindi devo fare l'iterazione sulle frasi. per i che va da 0 alla lunghezza della lista -1, quindi 
        #al numero di frasi contenute in quella lista.
        words = xml_dict['treebank']['sentence'][i]['word'] #words sono tutte le liste di parole delle frasi. words è una lista di dizionari che va in pasto 
        #alla funzione dataframe per trasformarlo in tabella. 

        # Verifica se 'words' è una lista o un singolo elemento. se non è una lista non funziona. 
        if isinstance(words, list): #isinstance è una funzione che controlla che la variabile words sia del tipo list
            xml_table = pd.DataFrame(words)
            #ora voglio selezionare e tenere solo alcune colonne della mia tabella, cioè lemma e pos, il resto non serve.
            xml_table = xml_table[['@ref','@id','@form','@lemma','@postag', '@relation']] #gli sto dicendo che il nuovo xml_table è uguale al vecchio, ma dove prendo solo
            #le colonne che interessano a me.
            xml_table['@id'] = list(range(1, len(xml_table)+1))  # <<<< Qui riscrivo gli ID consecutivi
        else:
            xml_table = pd.DataFrame([words]) #serve per trasformare quello che non è lista in una lista e poi trasforma words in una tabella. 
            xml_table = xml_table[['@ref','@id','@form','@lemma','@postag', '@relation']] #gli sto dicendo che il nuovo xml_table è uguale al vecchio, ma dove prendo solo
            #le colonne che interessano a me.
            xml_table['@id'] = [1]
        
        xml_table_final_listrelation[xml_file].append(xml_table) #nel passaggio prima ho creato la tabella di una frase di un xml.
        #ora questa tabella la devo mettere nella sua lista che è il valore dell'xml. scrivo il dizionario dell'xml corrente.append che è il metodo 
#che mi fa aggiungere un elemento in coda a una lista e così aggiungo la tabella nuova a fine lista. 
```
