```python
xml_list = os.listdir('C:/Users/lcolo/Desktop/TESI/CODICI DEFINITIVI DIZ UNICO-2/trasformazione degli xml in dataframe/glaux-main/glaux-main/xml') #nel corpus pedalion comparivano anche testi di poesia, ma a me interessa solo la prosa, quindi ho usato il metodo di libreria remove per togliere quelli che non mi servivano.
xml_table_final_listrelation = {}

for xml_file in xml_list:
    xml_table_final_listrelation[xml_file] = []
    with open('C:/Users/lcolo/Desktop/TESI/CODICI DEFINITIVI DIZ UNICO-2/trasformazione degli xml in dataframe/glaux-main/glaux-main/xml/' + xml_file, 'r', encoding='utf-8') as f:
        data = f.read()
    xml_dict = xmltodict.parse(data)

    # ðŸ‘‡ FORZA A LISTA LE FRASI
    sentences = xml_dict['treebank']['sentence']
    if not isinstance(sentences, list):
        sentences = [sentences]

    for i in range(len(sentences)):
        try:
            words = sentences[i]['word']

            if isinstance(words, list):
                xml_table = pd.DataFrame(words)
            else:
                xml_table = pd.DataFrame([words])
        except:
            print(xml_file)
        xml_table_final_listrelation[xml_file].append(xml_table)
    

```

```python
final_dic  = {k: pd.concat(v, ignore_index = True) for k ,v, in xml_table_final_listrelation.items()}
```

```python
tabelle_finali = {}

for xml_file in xml_list:
            
            final_dic[xml_file]['@id'] = range(1, len(final_dic[xml_file])+1)# <<<< Qui riscrivo gli ID consecutivi
            #ora voglio selezionare e tenere solo alcune colonne della mia tabella, cioÃ¨ lemma e pos, il resto non serve.
            
            xml_table = final_dic[xml_file][['@id','@form','@lemma','@postag', '@relation']] #gli sto dicendo che il nuovo xml_table Ã¨ uguale al vecchio, ma dove prendo solo
            #le colonne che interessano a me.
            
            tabelle_finali[xml_file] = xml_table
```

```python
for xml_file in xml_list:
    if xml_file in tabelle_finali:
        tabelle_finali[xml_file].to_csv('C:/Users/lcolo/Desktop/TESI/CODICI DEFINITIVI DIZ UNICO-2/GLAUX/output_glaux'+'/' + xml_file +'.csv', index=False)
        
```
